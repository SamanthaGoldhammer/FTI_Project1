# Global AI Content Impact

```
import pandas as pd
```

```
url = "https://raw.githubusercontent.com/frontiertechinstitute/datasets/refs/heads/main/AI%20and%20Digital%20Media/Global_AI_Content_Impact_Dataset.csv"
df = pd.read_csv(url)
df.head()
```

```
df.columns=(df.columns.str.strip())
df.columns=(df.columns.str.lower())
df.columns
```

```
df["country"]=df["country"].replace({"USA":"United States of America"})
```

```
df["country"]=df["country"].replace({"UK":"United Kingdom"})
```

```
automotive_industry = df[df['industry'] == 'Automotive']
print(automotive_industry[['country', 'revenue increase due to ai (%)','year']])
```

```
marketshareunder20 = df[df['market share of ai companies (%)'] < 20]
print(marketshareunder20[['country', 'revenue increase due to ai (%)', 'market share of ai companies (%)','year']])
```

```
marketshareover20_data = df[df['market share of ai companies (%)'] > 20]
print(marketshareover20_data[["consumer trust in ai (%)", "country", "year", "revenue increase due to ai (%)"]])
```

```
import matplotlib.pyplot as plt

revenue_df = df.groupby('country')['revenue increase due to ai (%)'].mean()
collaboration_df = df.groupby('country')['human-ai collaboration rate (%)'].mean()
trust_df = df.groupby('country')['consumer trust in ai (%)'].mean()

plt.figure(figsize=(12, 6))
plt.bar(revenue_df.index, revenue_df.values, color='skyblue')
plt.title('Average Revenue Increase Due to AI (%) by Country')
plt.xlabel('Country')
plt.ylabel('Revenue Increase Due to AI (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.bar(collaboration_df.index, collaboration_df.values, color='lightcoral')
plt.title('Average Human-AI Collaboration Rate (%) by Country')
plt.xlabel('Country')
plt.ylabel('Human-AI Collaboration Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.bar(trust_df.index, trust_df.values, color='lightgreen')
plt.title('Average Consumer Trust in AI (%) by Country')
plt.xlabel('Country')
plt.ylabel('Consumer Trust in AI (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
from scipy.stats import pearsonr
import numpy as np

selected_cols = ['consumer trust in ai (%)', 'revenue increase due to ai (%)', 'human-ai collaboration rate (%)']

for i in range(len(selected_cols)):
    for j in range(i + 1, len(selected_cols)):
        col1 = selected_cols[i]
        col2 = selected_cols[j]

        temp_df = df[[col1, col2]].dropna()

        if not temp_df.empty and len(temp_df) > 1:
            corr, p_value = pearsonr(temp_df[col1], temp_df[col2])
            print(f"P-value for correlation between '{col1}' and '{col2}': {p_value:.3f}")
        else:
            print(f"Could not calculate correlation for '{col1}' and '{col2}' due to insufficient data.")
```

```
correlation_matrix = df[['consumer trust in ai (%)', 'revenue increase due to ai (%)', 'human-ai collaboration rate (%)', 'ai adoption rate (%)', 'job loss due to ai (%)']].corr()
display(correlation_matrix)
```

```
from scipy import stats

strict_regulation_trust = df[df['regulation status'] == 'Strict']['consumer trust in ai (%)']
moderate_regulation_trust = df[df['regulation status'] == 'Moderate']['consumer trust in ai (%)']

t_statistic, p_value = stats.ttest_ind(strict_regulation_trust, moderate_regulation_trust)

print(f"Strict vs. Moderate")
print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

print(f"\nMean Consumer Trust (Strict Regulation): {strict_regulation_trust.mean():.2f}")
print(f"Mean Consumer Trust (Moderate Regulation): {moderate_regulation_trust.mean():.2f}")
```

```
from scipy import stats

strict_regulation_collaboration = df[df['regulation status'] == 'Strict']['human-ai collaboration rate (%)']
moderate_regulation_collaboration = df[df['regulation status'] == 'Moderate']['human-ai collaboration rate (%)']

t_statistic_collaboration, p_value_collaboration = stats.ttest_ind(strict_regulation_collaboration, moderate_regulation_collaboration)

print(f"T-statistic: {t_statistic_collaboration:.3f}")
print(f"P-value: {p_value_collaboration:.3f}")

print(f"\nMean Human-AI Collaboration Rate (Strict Regulation): {strict_regulation_collaboration.mean():.2f}")
print(f"Mean Human-AI Collaboration Rate (Moderate Regulation): {moderate_regulation_collaboration.mean():.2f}")
```

```
import matplotlib.pyplot as plt

strict_regulation_collaboration = df[df['regulation status'] == 'Strict']['human-ai collaboration rate (%)']
moderate_regulation_collaboration = df[df['regulation status'] == 'Moderate']['human-ai collaboration rate (%)']
mean_strict = strict_regulation_collaboration.mean()
mean_moderate = moderate_regulation_collaboration.mean()

regulation_labels = ['Strict Regulation', 'Moderate Regulation']
collaboration_means = [mean_strict, mean_moderate]

plt.figure(figsize=(5, 6))
plt.bar(regulation_labels, collaboration_means, color=['skyblue', 'lightcoral'])
plt.title('Average Human-AI Collaboration Rate by Regulation Status')
plt.xlabel('Regulation Status')
plt.ylabel('Human-AI Collaboration Rate (%)')
plt.ylim(20, 80)
plt.show()
```

```
plt.figure(figsize=(8, 6))
plt.boxplot([strict_regulation_collaboration, moderate_regulation_collaboration], tick_labels=regulation_labels)
plt.title('Human-AI Collaboration Rate Distribution by Regulation Status')
plt.xlabel('Regulation Status')
plt.ylabel('Human-AI Collaboration Rate (%)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
